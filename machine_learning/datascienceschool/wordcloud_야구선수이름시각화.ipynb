````python
from wordcloud import WordCloud
from collections import Counter
from PIL import Image
import sys
````

#### 야구선수 이름 리스트 만들기

````python
df = pd.read_csv('0409_data.csv')
name_list = list(df.이름)
````

#### 리스트에서 선수이름의 수를 세서 딕셔너리로 만들기

````python
counts = Counter(name_list)
names = dict(counts)
````

#### wordcloud를 넣을 이미지 불러오기

````python
baseball = np.array(Image.open('./6423096.png'))
plt.figure(figsize=(8, 8), dpi=70)
plt.imshow(baseball, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off'); plt.show()
````

#### wordcloud 시각화

````python
wc = WordCloud(font_path='/System/Library/Fonts/Supplemental/AppleGothic.ttf', 
               background_color='white', max_words=200, mask=baseball, stopwords={','}, max_font_size=150)
wc = wc.generate_from_frequencies(names)
# wc.words_

plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('baseball.png', dpi=300) # .png 파일로 저장
plt.show()
````
