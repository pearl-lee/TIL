```python
np.random.seed(1)
x = np.random.normal(size=1000)
x = np.hstack([x, 5 * np.ones(50)])
bins = np.linspace(-6, 6, 12 * 4 + 1)
ns, _ = np.histogram(x, bins=bins)

sample_mean = np.mean(x)
sample_median = np.median(x)
mode_index = np.argmax(ns)
sample_mode = 0.5 * (bins[mode_index] + bins[mode_index + 1])

sns.distplot(x, bins=bins)
plt.axvline(sample_mean, c='r', ls=':', label='표본평균')
plt.axvline(sample_median, c='g', ls='--', label='표본중앙값')
plt.axvline(sample_mode, c='m', ls='-', label='표본최빈값')
plt.title('표본평균, 표본중앙값, 표본최빈값의 차이')
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
data = df[df.species == 'virginica']['petal length (cm)']
bins = np.linspace(0, 10, 11)
ns, _ = np.histogram(data, bins=bins)

s_mean = np.mean(data)
s_median = np.median(data)
s_mode = (bins[np.argmax(ns)] + bins[np.argmax(ns) + 1]) * 0.5

sns.distplot(data, bins=bins)
plt.axvline(s_mean, c='r', ls=':', label='표본평균')
plt.axvline(s_median, c='g', ls='--', label='표본중앙값')
plt.axvline(s_mode, c='m', ls='-', label='표본최빈값')
plt.title('virginica 종 꽃잎길이의 표본평균, 표본중앙값, 표본최빈값의 차이')
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
data = df[df.species == 'virginica']['petal length (cm)']

def sample_statistics(data, n, x_min=0, x_max=10):
    bins = np.linspace(x_min, x_max, n+1)
    ns, _ = np.histogram(data, bins=bins)
    
    s_mean = np.mean(data)
    s_median = np.median(data)
    s_mode = (bins[np.argmax(ns)] + bins[np.argmax(ns) + 1]) * 0.5
    
    print('구간이 {}개:'.format(n))
    print('표본평균: {}\n표본중앙값: {}\n표본최빈값: {}'.format(s_mean, s_median, s_mode))

    sns.distplot(data, bins=bins)
    plt.axvline(s_mean, c='r', ls=':', label='표본평균')
    plt.axvline(s_median, c='g', ls='--', label='표본중앙값')
    plt.axvline(s_mode, c='m', ls='-', label='표본최빈값')
    plt.title('virginica 종 꽃잎길이의 표본평균, 표본중앙값, 표본최빈값의 차이')
    plt.xlabel('x')
    plt.legend()
    plt.show()
```

```python
sample_statistics(data, 50)
```

```python
sample_statistics(data, 100)
```

#### 연습문제 7.1.3
다우지수의 일간수익률데이터

(1) 다우지수 일간수익률의 표본평균, 표본분산, 표본표준편차를 구하라.

(2) 일간수익률의 표본표준편차에 대략적인 연간 거래일수 256일의 제곱근인 16을 곱해서 % 단위로 나타낸 것을 연간변동성(annual volatility)이라고 한다. 다우지수의 연간변동성을 구하라.

```python
import pandas_datareader.data as web
df = web.DataReader("DJCA", data_source='fred').dropna()['DJCA']
r = np.log(df / df.shift(1)).dropna().values

sample_statistics(r, 100, -0.09, 0.09)
```

```python
print('표본평균: {},\n표본분산: {},\n표본표준편차: {}'.format(np.mean(r), np.var(r), np.std(r)))
```

```python
print(f'표본평균: {np.mean(r)},\n표본분산: {np.var(r)},\n표본표준편차: {np.std(r)}')
# format 다른방법으로 쓰는 방법
```

```python
print('연간변동성: {} %'.format(np.std(r) * np.sqrt(256) * 10000))
```
