#### 주피터노트북에 출력되는 그래프 사이즈 조정하기

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot(dpi):
    fig, ax = plt.subplots(dpi=dpi)
    ax.plot([2, 4, 1, 5], label='Label')
    ax.legend()
plot(100)    
```

```python
fig, ax = plt.subplots(dpi=50)
ax.plot([[0, 0], [1, 1]], label='Label')
ax.legend()
```
