```python
from itertools import chain, combinations

def get_set_of_subsets(omega):
    return set([frozenset(s) for s in chain.from_iterable(combinations(omega, r) for r in range(len(omega) + 1))])
```

```python
omega2 = {'\u2660', '\u2663', '\u2665', '\u25C6'}
S = get_set_of_subsets(omega2)
S
```

#### 연습문제 6.2.6

주사위 확률 할당

```python
omega_dice ={1, 2, 3, 4, 5, 6}
SS = get_set_of_subsets(omega_dice)
SS
P_dice = {}

for i in SS:
    probability = 0.0
    if 1 in i:
        probability += 0.5
    if 2 in i:
        probability += 0.1
    if 3 in i:
        probability += 0.2
    if 4 in i:
        probability += 0.1
    if 5 in i:
        probability += 0.1
    if 6 in i:
        probability += 0
        
    P_dice[i] = probability

P_dice
```

0~180도에 화살이 2배 더 잘 박히도록 조작된 원반에 대하여, 
구간의 시작과 끝점을 입력받아 확률을 출력하는 함수 `P(a, b)` 구하기

```python
def P(a, b):
    if a > b:
        raise ValueError("a must be less than b of equal to b")
    a = np.maximum(0, a)
    b = np.minimum(b, 360)
    
    if b < 180:
        return (2/3)*((b-a)/100)
    else:
        if a < 180:
            return (2/3)*((180 - a)/180) + (1/3)*((b - 180)/180)
        return (1/3)*((b - a)/180)
P(30, 90)
```

```python
t = np.linspace(-100, 500, 300)
F = t / 360 
F[t < 180] = 2/3
F[t > 180] = 1/3
F[(t < 0) | (t > 360)] = 0 
plt.plot(t, F)
plt.ylim((-0.1, 1.1))
plt.xticks([0, 180, 360])
plt.title("질량밀도함수")
plt.show()
```

0~180도에 화살이 2배 더 잘 박히도록 조작된 원반에 대하여 누적분포함수 `F(x)` 구하기

```python
def F(x):
    if x < 0:
        return 0
    if x > 360:
        return 1
    elif x < 180:
        return (2 / 3) * (x / 180)
    elif x >= 180:
        return (2 / 3) + (1 / 3) * ((x - 180) / 180)
    

t = np.linspace(-100, 500, 100)
Fl = np.array(list(map(F, t)))
plt.plot(t, Fl)
plt.ylim((-0.1, 1.1))
plt.xticks([0, 180, 360])
plt.title("누적분포함수")
plt.show() 
```

조작되지 않은 원반의 확률밀도함수

```python
t = np.linspace(-100, 500, 1000)
F = t/360
F[t < 0] = 0
F[t > 360] = 1
p = np.gradient(F, 600/1000)

plt.plot(t, p)
plt.ylim((-0.0001, p.max()*1.1))
plt.xticks([0, 180, 360])
plt.title("조작되지 않은 원반의 확률밀도함수")
plt.show() 
```
